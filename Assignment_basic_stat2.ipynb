```python
# Sample Standard Deviation: Used when you have a subset of data from a larger group. It adjusts for having less data to give a better estimate of the overall variability.
# Population Standard Deviation: Used when you have all the data from an entire group. It shows how much the data varies from the average.

# Calculate Sample Standard Deviation:
# Step 1: Find the mean.
# Step 2: Subtract the mean from each score.
# Step 3: Square each deviation.
# Step 4: Add the squared deviations.
# Step 5: Divide the sum by one less than the number of data points.
# Step 6: Take the square root of the result from Step 5.

# Calculate population standard deviation:
# Step 1: Calculate the mean of the data
# Step 2: Subtract the mean from each data point. 
# Step 3: Square each deviation to make it positive.
# Step 4: Add the squared deviations together.
# Step 5: Divide the sum by the number of data points in the population. The result is called the variance.
# Step 6: Take the square root of the variance to get the standard deviation.
```

```python
# Load the dataset into a data analysis tool or programming environment 
import pandas as pd
import numpy as np
import operator
# Import math Library
import math
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29] 
```

```python
# Calculate Sample Standard Deviation:

# Importing the statistics module
import statistics as st

from scipy import stats

# Calculate the size of your sample (the number of measurements taken)
sample_size=len(data)

# Calculate μ — the mean (average) of the measurements
sample_mean =st.mean(data)

# The standard deviation of the measurements
sample_std_dev =st.stdev(data)

print ("Number of Samples:   ", sample_size)
print ("mean of the samples: ", sample_mean)
print ("standard deviation:  ", sample_std_dev)
```

```python
import scipy.stats as stats
import math

confidence_level = 0.99

df = sample_size - 1

alpha = (1 - confidence_level) / 2

t_value = stats.t.ppf(1 - alpha, df)

margin_of_error = t_value * (sample_std_dev / math.sqrt(sample_size))

lower_limit1 = sample_mean - margin_of_error
upper_limit1 = sample_mean + margin_of_error

print(f"Confidence Interval: ({lower_limit1}, {upper_limit1})")
```

```python
# Population Standard Deviation:
# Calculate the size of your sample (the number of measurements taken)
Population_size=len(data)

# Calculate μ — the mean (average) of the measurements
Population_mean =st.mean(data)

# The standard deviation of the measurements
Population_std_dev = 0.2 # Hard coded

print ("Number of Samples:   ", Population_size)
print ("mean of the samples: ", Population_mean)
print ("standard deviation:  ", Population_std_dev)
```

```python
import scipy.stats as stats
import math

confidence_level = 0.99

df = sample_size - 1

alpha = (1 - confidence_level) / 2

t_value = stats.t.ppf(1 - alpha, df)

margin_of_error = t_value * (Population_std_dev / math.sqrt(Population_size))

lower_limit2 = Population_mean - margin_of_error
upper_limit2 = Population_mean + margin_of_error

print(f"Confidence Interval: ({lower_limit2}, {upper_limit2})")
```

```python
# Plot confidence_interval for both std dev 
def plot_confidence_interval(x, values, z=1.96, color='#2187bb', horizontal_line_width=0.25):
    mean = statistics.mean(values)
    stdev = statistics.stdev(values)
    confidence_interval = z * stdev / sqrt(len(values))

    left = x - horizontal_line_width / 2
    top = mean - confidence_interval
    right = x + horizontal_line_width / 2
    bottom = mean + confidence_interval
    plt.plot([x, x], [top, bottom], color=color)
    plt.plot([left, right], [top, top], color=color)
    plt.plot([left, right], [bottom, bottom], color=color)
    plt.plot(x, mean, 'o', color='#f44336')

    return mean, confidence_interval

plt.xticks([1, 2], ['Sample std dev', 'Population std dev'])
plt.title('Confidence Interval')
plot_confidence_interval(1, [lower_limit1, upper_limit1])
plot_confidence_interval(2, [lower_limit2, upper_limit2])

plt.show()
```
