```python
# Load the dataset into a data analysis tool or programming environment 
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\DataScience\Assignments\01 Basic stats - 1\Basic stats - 1\sales_data_with_discounts.csv")
df
```

```python
# Calculate the mean, median, mode, and standard deviation for these columns.
import statistics as st

# For Volume: Calculate medin, mean, Variance and standard devation 
mean_Valume =st.mean(df.Volume)
median_Valume =st.median(df.Volume)
mode_Valume =st.multimode(df.Volume)
variance_Valume =st.variance(df.Volume)
stdev_Valume =st.stdev(df.Volume)
print("Valume: ")
print("mean    ", mean_Valume)
print("median  ", median_Valume)
print("mode    ", mode_Valume)
print("variance", variance_Valume)
print("stdev:  ", stdev_Valume)

# brief interpretation of these statistics:

# mean: Avarage value of this column is 5.05
# median: Middle value this column is 4.0 after arranging in oder
# mode: Only one value i.e 10 is repative (frequently occurring) in this column
# variance: 17.9 i.e average of "squared deviation scores" where "squared deviation scores" is observation - true mean
# standard deviation: 4.23 i.e square root of variance

```

```python
# Step 1:
# Load the dataset into a data analysis tool or programming environment 
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\DataScience\Assignments\01 Basic stats - 1\Basic stats - 1\sales_data_with_discounts.csv")
df
```

```python
df.columns
```

```python
# Step 2: Calculate the Mean
# Calculate the mean, median, mode, and standard deviation for these columns.
import statistics as st

# For Volume: Calculate mean and standard devation 
mean_Valume =st.mean(df.Volume)
stdev_Valume =st.stdev(df.Volume)

# Perform Z-score normalization
z_scores_Volume = (df.Volume - mean_Valume) / stdev_Valume

# Print the results
print("Original data:", df.Volume)
print("Mean:", mean_Valume)
print("Standard Deviation:", stdev_Valume)
print("Z-scores:", z_scores_Volume)
```

```python
#Step 3 : Outlier detection
# Outlier detection: Z-scores beyond Â±3 are considered outliers
outliers = np.where(np.abs(z_scores_Volume) > 3)
print (outliers)
#plt.plot(outliers[0], df.Volume[outliers], 'ro', label='Outliers')
plt.figure(figsize=(20, 8))
plt.plot(df.Volume, 'bo-', label='Original Data')
#plt.plot(outliers[0], df.Volume[outliers], 'ro', label='Outliers')
plt.axhline(mean_Valume, color='g', linestyle='--', label='Mean')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Data with Outliers Detected')
plt.legend()
plt.show()
```
